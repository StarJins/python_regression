```python
data = pd.read_csv("C:/Users/Jintae/Desktop/python_regression/YEAR_DATA.csv", encoding='CP949')
print(data.head(10))
```

```python
data_c = data[data.업종 == '피자']
print(data_c)
```

```python
#data2 = pd.get_dummies(data_jok, columns=['요일', '시간대'], prefix=['요일', '시간대'])
#print(data2.head(10))

data2 = pd.get_dummies(data_c, columns=['일자', '요일', '시간대', '업종', '시도', '시군구', '읍면동'])
#print(list(data2.columns))
```

```python
features = data2.loc[:, '일자_20180101':'읍면동_흥인동']

X = features.values
y = data2['통화건수'].values

print("X.shape : {}  y.shape : {}".format(X.shape,y.shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

model1 = linear_model.LinearRegression()
model1.fit(X_train, y_train)
#print(model1.coef_, model1.intercept_)
```

```python
print ("Model1 Training Mean squared error: %.2f" % np.mean((model1.predict(X_train) - y_train) ** 2))
#print ("Model1 Test Mean squared error: %.2f" % np.mean((model1.predict(X_test) - y_test) ** 2))
print('모델 train 점수: {:.2f}'.format(model1.score(X_train, y_train)))
```
