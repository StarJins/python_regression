```python
"""
참고 사이트

sklearn 사용한 선형회귀분석
https://3months.tistory.com/137
https://datascienceschool.net/view-notebook/58269d7f52bd49879965cdc4721da42d/
https://cyan91.tistory.com/39

범주형 데이터 처리 후 sklearn 사용 방법
http://blog.naver.com/PostView.nhn?blogId=ssdyka&logNo=221284732634&parentCategoryNo=&categoryNo=&viewDate=&isShowPopularPosts=false&from=postView
"""

import pandas as pd
from sklearn import datasets, linear_model, preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("C:/Users/Jintae/Desktop/python_regression/YEAR_DATA.csv", encoding='CP949')
print(data.head(10))
```

```python
data_c = data[data.업종 == '피자']
print(data_c)
```

```python
#data2 = pd.get_dummies(data_jok, columns=['요일', '시간대'], prefix=['요일', '시간대'])
#print(data2.head(10))

#  one-hot encoding을 한 후에는 columns 하나를 drop 해야한다. 독립변수의 독립성 보장을 위해
data2 = pd.get_dummies(data_c, columns=['일자'], drop_first=True)
data2 = pd.get_dummies(data2, columns=['요일'], drop_first=True)
data2 = pd.get_dummies(data2, columns=['시간대'], drop_first=True)
print(list(data2.columns))
```

```python
features = data2.loc[:, '일자_20180102':'시간대_23']
#print(features)

X = features.values
y = data2['통화건수'].values

print("X.shape : {}  y.shape : {}".format(X.shape,y.shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

model1 = linear_model.LinearRegression()
model1.fit(X_train, y_train)
#print(model1.coef_, model1.intercept_)
```

```python
print ("Model1 Training Mean squared error: %.2f" % np.mean((model1.predict(X_train) - y_train) ** 2))
#print ("Model1 Test Mean squared error: %.2f" % np.mean((model1.predict(X_test) - y_test) ** 2))
print('모델 train 점수: {:.2f}'.format(model1.score(X_train, y_train)))
```
